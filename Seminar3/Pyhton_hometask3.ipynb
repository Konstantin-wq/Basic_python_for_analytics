### 1.1 Считать данные с помощью pandas. Вывести на экран первые 5 строк

```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv',header = 0) #Считываем файл csv
df.head() #Выводим первые 5 строк
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.  Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built'] #ВЫчисляем разницу между годом реновации и годом постройки
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: x if x >0 else 0)#Заменяем отрицательные значения на '0'
df.head(10)
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['date']= pd.to_datetime(df['date']) #Преобразуем столбец 'date' в формат datetime
df['year_sale'] = pd.DatetimeIndex(df['date']).year #Получаем год
df['month_sale'] = pd.DatetimeIndex(df['date']).month # Получаем месяц
df['month_name'] = pd.DatetimeIndex(df['date']).month_name() #Получаем название месяца
df.head(10)
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop( columns = ['date','zipcode','lat','long'],inplace = True)
df.head(10)
```

## Задача 3 Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price':'mean'}).sort_values('price') #1 метод
```

```python
df.pivot_table(index='bedrooms', values = 'price', aggfunc = 'mean').sort_values('price') #2 метод
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price':['min','max','mean']}) #1 метод
```

```python
df.pivot_table(index = 'condition', values = 'price', aggfunc =['min','max','mean']).sort_values('condition') #2 метод
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.groupby(['waterfront','view']).agg({'id':'count'}).unstack(-1).fillna(0) #1 вариант
```

```python
df.pivot_table(index = 'waterfront',columns = 'view',values = 'id', aggfunc = 'count' ).fillna(0) #2 вариант
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index = df['floors'], columns = df['bedrooms'],margins=True, margins_name='Total') #1 вариант
```

```python
df.pivot_table(index = 'floors',columns = 'bedrooms', values = 'id', aggfunc = 'count',margins=True, margins_name='Total').fillna(0) #2 вариант
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.groupby(['condition','grade']).agg({'price':'median'}).unstack(-1).fillna(0) #1 вариант
```

```python
df.pivot_table(index = 'condition', columns = 'grade', values = 'price', aggfunc = 'median').fillna(0) #2 вариант
```
